```python
import os
import cv2
import matplotlib.pyplot as plt

# Ambil folder hasil deteksi terbaru
detect_path = 'runs/detect'
exp_folders = sorted(os.listdir(detect_path), key=lambda x: os.path.getmtime(os.path.join(detect_path, x)))
latest_folder = exp_folders[-1]
latest_path = os.path.join(detect_path, latest_folder)

# Ambil nama file gambar deteksi
result_images = [f for f in os.listdir(latest_path) if f.endswith(('.jpg', '.png'))]
latest_result = os.path.join(latest_path, result_images[0])

# Tampilkan hasil
img = cv2.imread(latest_result)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8, 8))
plt.imshow(img)
plt.axis('off')
plt.title('HASIL DETEKSI TERBARU')
plt.show()
```
